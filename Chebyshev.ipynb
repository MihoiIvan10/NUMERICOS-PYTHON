```python
# NODOS DE Chebyshev

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

# Función de Runge
def runge(x):
    return 1 / (1 + 25 * x**2)

def chebyshev_nodes(a, b, n):
    nodes = np.cos((2 * np.arange(1, n + 1) - 1) * np.pi / (2 * n))
    return 0.5 * (a + b) + 0.5 * (b - a) * nodes


a, b = -1, 1
n_values = [7, 9, 11, 25]


x_plot = np.linspace(a, b, 500)
y_runge = runge(x_plot)

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_runge, 'k-', label='Función de Runge', linewidth=2)


for n in n_values:
    # Nodos de Chebyshev
    nodes = chebyshev_nodes(a, b, n)
    y_nodes = runge(nodes)

    # Polinomio de Lagrange
    poly = lagrange(nodes, y_nodes)
    y_interp = poly(x_plot)

    # Graficar
    plt.plot(x_plot, y_interp, '--', label=f'n = {n}', linewidth=1.5)
    plt.plot(nodes, y_nodes, 'o', markersize=5)

plt.title('Interpolación de la función de Runge usando nodos de Chebyshev')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
