```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Funcin de Taylor truncada 
def obtener_serie_taylor(funcion, variable, punto_expansion, orden):
    return sp.series(funcion, variable, punto_expansion, orden).removeO()

# Funcin para convertir una expresin simblica en una funcin numrica
def convertir_a_funcion_numerica(serie_taylor, variable):
    return sp.lambdify(variable, serie_taylor, "numpy")

# Funcin ln(x)
def f_original(x):
    return np.log(x)

# Funcin para crear el grfico de la funcin y la serie de Taylor
def graficar(funcion_original, funcion_taylor, x_vals):
    f_vals = funcion_original(x_vals)
    taylor_vals = funcion_taylor(x_vals)

    plt.figure(figsize=(8, 6))
    plt.plot(x_vals, f_vals, label=r'$\ln(x)$', color='blue', lw=2)
    plt.plot(x_vals, taylor_vals, label='Serie de Taylor (orden 4)', color='red', linestyle='dashed', lw=2)
    plt.title('Comparacin de ln(x) y su Serie de Taylor (orden 4)', fontsize=14)
    plt.xlabel('x', fontsize=12)
    plt.ylabel('f(x)', fontsize=12)
    plt.legend()
    plt.grid(True)
    plt.show()


x = sp.symbols('x')
f = sp.ln(x)

# Obtener la serie de Taylor truncada de f(x) alrededor de x0 = 1, truncada a cuarto orden
taylor_series = obtener_serie_taylor(f, x, 1, 5)


print(f"Serie de Taylor truncada de orden 4 de ln(x) alrededor de x0 = 1:\n{taylor_series}")


taylor_func = convertir_a_funcion_numerica(taylor_series, x)


x_vals = np.linspace(0.1, 3, 400)


graficar(f_original, taylor_func, x_vals)
```
